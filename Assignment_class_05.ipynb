```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
m = MinMaxScaler()
```

```python
df['Marketing Spend'] = m.fit_transform(df[['Marketing Spend']])
```

```python
df['Administration'] = m.fit_transform(df[['Administration']])
```

```python
df['Transport'] = m.fit_transform(df[['Transport']])
```

```python
df['Profit'] = m.fit_transform(df[['Profit']])
```

```python
df.head()
```

```python
import numpy as np
for x in df1:
    if df1.dtypes[x] == np.float64:
        df1[x] = m.fit_transform(df1[[x]])
    else:
        pass


df1.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df2.head()
```

```python
for x in df2.columns:
    if df2[x].dtype == np.float64:
        df2[x] = sc.fit_transform(df2[[x]])
    else:
        continue

df2.head()
```

# Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
```

```python
rc = RobustScaler()
```

```python
df3.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for x in df3.columns:
    if is_numeric_dtype(df3[x]):
        df3[x] = rc.fit_transform(df3[[x]])
    else:
        continue
df3.head()
```

# Max Absolute Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
mas = MaxAbsScaler()
```

```python
df4.head()
```

```python
for x in df4.columns:
    if is_numeric_dtype(df4[x]):
        if x != 'Profit':       # Y variable does not need to be transformed   
            df4[x] = mas.fit_transform(df4[[x]])
        else:
            continue
    else:
        continue
df4.head()
```
